```python
dados1 = dados1.drop(columns=['SG_UF_NASCIMENTO', 'TP_ANO_CONCLUIU','TP_ENSINO','CO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC'],)
```

```python
dados1.head()
```

```python
dados1 = dados1.drop(columns=['CO_UF_RESIDENCIA', 'CO_UF_NASCIMENTO'],)
```

```python
dados1.dtypes
```

```python
dados1.loc[:,'NU_NOTA_CH']/=10
dados1.loc[:,'NU_NOTA_LC']/=10
dados1.loc[:,'NU_NOTA_MT']/=10
dados1.loc[:,'NU_NOTA_CN']/=10
```

```python
dados1.head()
```

###Renomeando Variáveis (Colunas)


```python
dados1 = dados1.rename(columns = {'NU_NOTA_REDACAO':'NOTA_REDACAO'})
```

```python
dados1.head()
```

```python
dados1 = dados1.rename(columns = {'NU_NOTA_CN':'NOTA_CN','NU_NOTA_CH':'NOTA_CH', 'NU_NOTA_LC':'NOTA_LC', 'NU_NOTA_MT':'NOTA_MT' })
```

```python
dados1 = dados1.rename(columns = {'NU_NOTA_COMP1':'COMP1','NU_NOTA_COMP2':'COMP2','NU_NOTA_COMP3':'COMP3',
                                  'NU_NOTA_COMP4':'COMP4','NU_NOTA_COMP5':'COMP5'})
```

```python
dados1 = dados1.rename(columns ={'NU_IDADE':'IDADE','TP_SEXO': 'SEXO',
                                  'TP_COR_RACA':'RACA', 'Q025': 'INTERNET',
                                  'TP_ESCOLA':'ESCOLA'})
```

```python
dados1 = dados1.rename(columns = {'NU_INSCRICAO':'INSCRICAO','NO_MUNICIPIO_RESIDENCIA':'MUNICIPIO','TP_ESTADO_CIVIL': 'ESTADO_CIVIL', 'TP_NACIONALIDADE':'NACIONALIDADE',})
```

###Renomeando Linhas

```python
dados1['RACA'] = dados1['RACA'].replace({0:'não_declarado', 1:'branca', 2:'preta', 3:'parda', 4:'amarela', 5:'indigena'})
```

```python
## OUTRA MANEIRA DE RENOMEAR LINHAS
# dados1['RACA'] = dados1['RACA'].replace([0,1,2,3,4,5],['Não Declarado, 'Branca', 'Preta', 'Parda', 'Amarela', 'Indígena'])
```

```python
dados1['TP_LINGUA'] = dados1['TP_LINGUA'].replace({0:'Inglês', 1:'Espanhol'})
```

```python
dados1['ESCOLA']= dados1['ESCOLA'].replace({1:'não_respondeu', 2:'pública', 3:'privada', 4:'exterior'})
```

```python
dados1['INTERNET'] = dados1['INTERNET'].replace({'A':'Não', 'B':'Sim'})
```

```python
dados1['TP_LINGUA'] = dados1['TP_LINGUA'].replace({'Espanhos': 'Espanhol'})
```

```python
dados1['ESTADO_CIVIL'] = dados1['ESTADO_CIVIL'].replace({0:'não_informado', 1:'solteiro', 2:'Casado/Mora com companheiro', 3:'Divorciado', 4:'Viúvo'})
```

```python
dados1['NACIONALIDADE'] = dados1['NACIONALIDADE'].replace({0:'não_informado',1:'brasileiro(a)',2:'brasileiro_naturalizado',3:'estrnageio',4:'brasileirot nato'})
```

```python
dados1['IN_TREINEIRO'] = dados1['IN_TREINEIRO'].replace({'Não': 0, 'Sim':1})
```

## Análise Exploratória e Transformação dos Dados

### Usando a função QUERY para comparar uma coluna com a outra.

```python
menores_12 = dados1.query('IDADE < 12')['MUNICIPIO'].value_counts( )
```

```python
# inscrição com idad abaixo de 12 anos serão excluídos devido a alta probabilidade de erro.
menores_12
```

```python
dados_maiores_11 = dados1.loc[dados1.IDADE > 11]
```

```python
dados_maiores_11
```

```python
dados_maiores_11['IDADE'].value_counts().sort_index()
```

### Treineiros 

```python
dados_maiores_11.query('IN_TREINEIRO == 1 ')['IN_TREINEIRO'].value_counts()
```

```python
treineiros = dados_maiores_11.loc[dados_maiores_11.IN_TREINEIRO == 1]
```

```python
treineiros.head()
```

```python
# Salvando treineiros com csv
treineiros.to_csv('treineiros_enem_2019_sp.csv',encoding = 'iso-8859-1', index = False)
```

```python
# Separaçao entre os treineiros e os vestibulandos
vestibulandos = dados_maiores_11.loc[dados_maiores_11.IN_TREINEIRO == 0]
```
